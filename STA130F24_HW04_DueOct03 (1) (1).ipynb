#### Standard Deviation:
The standard deviation measures the spread of individual data points around the mean in a dataset. It tells us how much individual data points differ from the mean. A high standard deviation indicates that data points are spread out over a wide range of values, while a low standard deviation indicates that the data points are clustered closely around the mean.

#### Standard Error of Mean:
The standard error of the mean shows the precision of the sample mean as an estimate of the population mean. It tells us how much variability we should expect in the sample means if we took multiple samples from the same population. It also reflects how confident we can be in the sample mean as a good representation of the population mean.

In the case of bootstrapping, after drawing a large number of bootstrap samples and calculating the mean of each, the standard error of the mean is calculated as the standard deviation of the distribution of the bootstrap means. This reflects how much the sample means vary across the different bootstrap samples.

### **1. Standard Error of the Mean (SEM) vs. Standard Deviation (SD)**:
- **SD** measures the variability of individual data points in a dataset around the mean. It tells you how much individual values differ from the average.
- **SEM** measures the variability of sample means from the population mean. It quantifies how much the sample mean would fluctuate if you took many samples from the same population. SEM is calculated as \( \text{SEM} = \frac{\text{SD}}{\sqrt{n}} \).
- **Key difference**: SD measures the spread of individual data points, while SEM measures the precision of the mean as an estimate of the population mean.

### **2. SD and SEM in Bootstrapping**:
- In bootstrapping, **SD** describes the variability within individual bootstrap samples or across bootstrap sample statistics (e.g., sample means).
- **SEM** is used in bootstrapping to describe the variability in the sample means and to estimate the precision of the mean estimate. It is the standard deviation of the distribution of bootstrap sample means.

### **3. Using SEM to Create a 95% Confidence Interval**:
- A **95% confidence interval** can be created using SEM by assuming the distribution of bootstrapped means is normal:
  \[
  \text{95% Confidence Interval} = \overline{X} \pm 1.96 \times \text{SEM}
  \]
  where \( \overline{X} \) is the sample mean, and \( 1.96 \) is the critical value for 95% confidence.
- This method assumes that the bootstrapped means follow a roughly normal distribution.

### **4. Creating a 95% Bootstrapped Confidence Interval Without SEM (Percentile Method)**:
- The **percentile method** uses the distribution of bootstrapped sample means without relying on SEM:
  - Generate a large number of bootstrap samples and compute the sample mean for each.
  - Sort the bootstrapped means and find the **2.5th percentile** and the **97.5th percentile**.
  - These percentiles form the **95% bootstrapped confidence interval**.
- This method is non-parametric and doesn’t assume a normal distribution, making it robust for non-normal or small datasets.

### **Example**:
If the 2.5th and 97.5th percentiles of the bootstrapped means are 45 and 55, respectively, then the 95% confidence interval is \( [45, 55] \).

````python
import numpy as np

# Function to create a bootstrap confidence interval
def bootstrap_ci(data, num_bootstrap=10000, ci_level=95):
    # Array to store the bootstrap means
    bootstrap_means = [] # We can change the variable's name to bootstrap_median if we want use median instead of mean
    
    # Resample and calculate the mean for each bootstrap sample
    for _ in range(num_bootstrap):
        # Resample with replacement from the original data
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        # if we want to use the median instead of mean, we can replace np.mean() with np.median()
        bootstrap_means.append(np.mean(bootstrap_sample))
    
    # Calculate the confidence interval
    lower_bound = np.percentile(bootstrap_means, (100 - ci_level) / 2)
    upper_bound = np.percentile(bootstrap_means, 100 - (100 - ci_level) / 2)
    
    return lower_bound, upper_bound

# Example usage
data = [12, 15, 14, 10, 13, 17, 12, 16, 11]  # Sample data
ci = bootstrap_ci(data)
print(f"95% Confidence Interval for the mean: {ci}")
````

#### 6.1
Bootstrapping is like making many small copies of the data by repeatedly selecting parts of it at random. Imagine there are 100 people in a class, and you randomly choose 40 of them to measure their heights. Now, you have a dataset with the heights of these 40 people. From this dataset, you randomly draw one height and then put it back; this is called sampling with replacement, meaning that the same height can be selected more than once. You repeat this process 40 times to create a new set of 40 heights. Then you repeat this entire process—randomly drawing 40 heights from the original set of 40 people's heights—multiple times. By generating a large number of these datasets, you can estimate something like the mean height of the entire class of 100 people using only the data from 40 of them.

2. **Confidence Interval Overlapping Zero**:
   - A confidence interval overlapping zero suggests that zero is a plausible value for the population parameter (e.g., mean difference). This overlap implies that **no effect** is possible based on the data, which leads to **failing to reject the null hypothesis**.
   
   - Even if the observed sample mean is not zero, if the confidence interval includes zero, the data does not conclusively rule out the possibility of no effect, so we do not reject the null hypothesis.

3. **Rejecting the Null Hypothesis**:
   - To reject the null hypothesis, the data must provide strong evidence that the true population parameter is significantly different from the null value (often zero).
   
   - You reject the null hypothesis when:
     1. The **confidence interval** does not overlap with zero, indicating a significant effect.
     2. The **p-value** is below the significance level (typically < 0.05).
     3. A larger sample size or a larger effect size strengthens the evidence.

### Key Points for Submitting:
- Null hypothesis: "no effect or difference."
- Failing to reject means the evidence is insufficient.
- Confidence interval including zero supports failing to reject the null hypothesis.
- Rejecting the null requires strong evidence: CI does not include zero, p-value < significance level, or a large sample size.

I would like to determine whether there is an effect on health improvement between the initial and final health scores of patients.

The null hypothesis is that the patients' final health scores are not significantly different from their initial health scores. This would imply no measurable health improvement or deterioration.

````python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
np.random.seed(120)
df = pd.read_csv("data.csv")
df.describe()
````

````python
df_melted = pd.melt(df, id_vars=['PatientID'], value_vars=['InitialHealthScore', 'FinalHealthScore'],
                    var_name='HealthScoreType', value_name='HealthScore')

# Create the box plot
fig = px.box(df_melted, x='HealthScoreType', y='HealthScore', title='Comparison of Initial and Final Health Scores')
fig.show(renderer="png")
````

From the above analysis, it is clear that overall, the patients' health scores increased. However, we still cannot be certain that the vaccine is effective, so we will perform bootstrapping next.

````python
df['HealthScoreDifference'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Number of bootstrap samples
n_iterations = 1000
n_sample_size = 10  # Bootstrap sample size
bootstrap_differences = []

# Extracting the differences
differences = df['HealthScoreDifference'].to_numpy()

# Bootstrapping process
for _ in range(n_iterations):
    # Resample using numpy with replacement
    bootstrap_difference = np.random.choice(differences, size=n_sample_size, replace=True)
    
    # Calculate the mean for the bootstrap sample
    mean_difference = bootstrap_difference.mean()
    
    # Append the mean difference to the list
    bootstrap_differences.append(mean_difference)

# Convert the list to a numpy array for easier manipulation
bootstrap_differences = np.array(bootstrap_differences)

# Plotting the histogram of the differences
plt.figure(figsize=(8, 6))
plt.hist(bootstrap_differences, bins=30, color='purple', alpha=0.7, edgecolor='black')
plt.title('Bootstrap Distribution of Health Score Differences')
plt.xlabel('Mean Difference (Final - Initial)')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linestyle='--', label='Mean Difference = 0')
plt.legend()
plt.show()

# Calculate the overall observed mean difference
observed_difference = differences.mean()
lower_bound = np.percentile(bootstrap_differences, 2.5)
upper_bound = np.percentile(bootstrap_differences, 97.5)

print(f'Observed Mean Difference (Final - Initial): {observed_difference:.2f}')
print(f'95% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]')
````

The above code performs a bootstrapping analysis to determine the effectiveness of a vaccine by comparing initial and final health scores. First, it creates a new column by calculating the difference between each patient's FinalHealthScore and InitialHealthScore. Then, it generates 1,000 bootstrap samples from the values in this new column and calculates the mean for each bootstrap sample. Finally, it plots all these bootstrap means in a histogram and calculates their overall mean and 95% confidence interval.

After bootstrapping, it is clear that the 95% confidence interval does not include zero, which indicates that the vaccine is effective in improving the health scores of patients.
