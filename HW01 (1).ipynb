## Question 7

### 7.1

### df.dropna() is Preferred over del df['col'] in the Following Scenarios
- A small portion of the data is missing a column, so dropping rows is more efficient than removing the entire columns.
- When there's multiple missing values in a row, dropping it is more practical than deleting columns.
- If you plan to use the column in other analysis, it's better to drop rows rather than deleting the entire column.

Below is an example:

````python
# Drop rows with missing 'age' values, but keep the 'age' column for future analysis
df2.dropna(subset=['age'])
````

### 7.2

### del df['col'] is Preferred over df.dropna() in the Following Scenarios
- When a column has a significant portion of it's data missing, discard the entire column is more efficient since keeping the column may not add much value to the analysis.
- When a column is not relevent to our analysis and any other analysis we plan to do, we can just remove it to optimize our memory.
- When a column contains only a single value (or nearly the same value across all rows). It doesn't provide any useful information for our analysis. 

Below is an example:

````python
# Drop the 'deck' column since it's mostly missing and not useful
del df2['deck']
df2
````

### 7.3

### Why Applying del df['col'] before df.dropna()
- If there is a column that has too much missing values when df.dropna() is applied, it could cause excessive row dropping which means rows with valid data may get removed just because they have missing data in a column that is irrelevant or already slated for deletion.
- By deleting the columns you don't need early on, it reduces the dimensionality of the dataframe which speed up computation.

### 7.4

### Remove All Missing Data from the Dataset using Combination of del df['col'] and/or df.dropna()

````python
df3 = pd.read_csv(url2)
df3
````

````python
# Calculate the percentage of missing values per column
df3.isna().mean()*100
````

````python
# Remove the column that has over 50% of it's data missing
del df3['deck']
df3
````

````python
# Now, we can remove any rows that still contain missing values in the remaining columns.
df3.dropna()
````

### 8.1

### Function of df.groupby("col1")["col2"].describe() 
- First, it splits the data into subsets/groups where each subset/group contains all rows that share the same value in "col1".
- Next, it specifies that the operation should be applied to column "col2" for each group from the previous step.
- Finally, .describe() calculates the statistics (like count, mean, std, min, quartiles, and max) for "col2" within each group created by "col1".

Below is an example from the "titanic" data set:

````python
df4.groupby("sex")["age"].describe()
````

````python
df4.describe(include='all')
````

### 8.2

### The Different Values of Count Results from df.describe() and df.groupby("col1")["col2"].describe()
- The count results from df.describe() is the number of non-missing values in that column. 
- The count results from df.groupby("col1")["col2"].describe() shows the number of non-missing values in "col2" for each group in "col1". 

This can be shown from the above example that the sum of count of male and female is 714 which is not the count for 'sex' from df.describe, but rather the count for 'age'. 

Therefore, the function of count in conjunction with group_by is to show the number of non-missing values in "col2" for each group in "col1".
