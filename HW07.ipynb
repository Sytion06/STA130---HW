### Question 1

1. Simple Linear Regression vs. Multiple Linear Regression
- Simple Linear Regression is a model with a single predictor (independent variable) and a response (dependent variable), modeling the relationship as:

  y=Œ≤0+Œ≤1x+œµ
  
  where y is the dependent variable, x is the single independent variable, beta0 is the intercept, beta1 is the coefficient for x, and epsilon is the error term.
  
- Multiple Linear Regression includes two or more predictors, allowing for a model like:

  y=Œ≤0+Œ≤1x1+Œ≤2x2+‚ãØ+Œ≤kxk+œµ
  
  This approach provides the benefit of incorporating multiple factors that may influence y, capturing more complex relationships and generally improving prediction accuracy by accounting for additional sources of variation.

2. Continuous vs. Indicator Variables in Simple Linear Regression
- Continuous Variables are numeric and measured on a continuous scale (e.g., age, height), which allows for regression coefficients that quantify changes in y per unit increase in the predictor.
- Indicator Variables take on values of 0 or 1 to represent categories (e.g., boolean expression as 0 for false and 1 for true). In simple linear regression, an indicator variable shifts the predicted outcome between two groups but does not model variation within the groups.

3. Model Behavior with an Indicator Variable Added in Multiple Linear Regression
- When a single indicator variable is added to a continuous predictor in a Multiple Linear Regression model, it effectively creates separate regression lines (intercepts) for each category represented by the indicator. This means the model can capture differences in the response variable between categories, while still using the continuous predictor for within-group variation.
  
  For example, the model:
  y=Œ≤0+Œ≤1x+Œ≤2D+œµ
  
  where D is an indicator variable (e.g., gender), would create two parallel lines (if no interaction term is added) with different intercepts for each category (e.g., male and female).

4. Adding an Interaction Between a Continuous and Indicator Variable in Multiple Linear Regression
- Adding an interaction term between a continuous variable and an indicator variable introduces separate slopes for each category of the indicator. The model form would look like:
  y=Œ≤0+Œ≤1x+Œ≤2D+Œ≤3(x‚ãÖD)+œµ
  
  where Œ≤3 represents the interaction effect. This setup means that the relationship between y and x changes based on the category of D, allowing the model to capture distinct patterns for each group.

5. Multiple Linear Regression with Only Indicator Variables from a Categorical Variable
- When using only indicator variables derived from a non-binary categorical variable (e.g., race with multiple categories) in a Multiple Linear Regression model, each category is typically encoded as a binary indicator in a process known as one-hot encoding. For a categorical variable with \(k\) categories, \(k-1\) binary indicators are created, with the remaining category serving as the reference.
  
  The model then takes the form:
  y=Œ≤0+Œ≤1D 1+Œ≤2D2+‚ãØ+Œ≤k‚àí1Dk‚àí1+œµ
  
  Here, each coefficient Œ≤i represents the difference in the response variable relative to the reference category. This allows the model to capture different mean values for each category but does not model relationships within each group (e.g., age effects within each racial group).

```python
import numpy as np
import plotly.graph_objects as go

# Simulate data range for the continuous predictor (e.g., WELLNESS_life_satisfaction)
np.random.seed(42)
n = 100
continuous_variable = np.linspace(1, 10, n)  # Simulate WELLNESS_life_satisfaction across its range
binary_variable = np.random.choice([0, 1], size=n)  # Simulate a binary indicator (e.g., WELLNESS_self_rated_mental_health)

# Extract coefficients from the logistic model output provided
intercept = -3.3281
slope_continuous = 0.3969  # Example coefficient for WELLNESS_life_satisfaction (mid-range estimate)
slope_binary = 0.3837  # Example coefficient for binary indicator (Fair mental health)
interaction_effect = 0.7032  # Example interaction effect for continuous * binary

# Generate response variables for both models
# Additive model: Response ~ Intercept + slope_continuous * Continuous Predictor + slope_binary * Binary Predictor + noise
random_noise = np.random.normal(0, 1, n)  # Random noise for realism in the data points
response_additive = intercept + slope_continuous * continuous_variable + slope_binary * binary_variable + random_noise

# Synergistic model: Response ~ Intercept + slope_continuous * Continuous Predictor + slope_binary * Binary Predictor + interaction_effect * (Continuous Predictor * Binary Predictor) + noise
response_synergistic = (intercept + slope_continuous * continuous_variable + 
                        slope_binary * binary_variable + interaction_effect * continuous_variable * binary_variable + 
                        random_noise)

# Best fit lines (calculated without noise) for the two models
# Additive model fit lines
fitted_line_additive_0 = intercept + slope_continuous * continuous_variable  # for Binary Indicator = 0
fitted_line_additive_1 = intercept + slope_continuous * continuous_variable + slope_binary  # for Binary Indicator = 1

# Synergistic model fit lines
fitted_line_synergistic_0 = intercept + slope_continuous * continuous_variable  # for Binary Indicator = 0
fitted_line_synergistic_1 = (intercept + slope_continuous * continuous_variable + 
                             slope_binary + interaction_effect * continuous_variable)  # for Binary Indicator = 1

# Prepare data for plotting
df_additive = pd.DataFrame({
    'Continuous Predictor': continuous_variable,
    'Binary Predictor': binary_variable,
    'Response': response_additive
})

df_synergistic = pd.DataFrame({
    'Continuous Predictor': continuous_variable,
    'Binary Predictor': binary_variable,
    'Response': response_synergistic
})

# Plot Additive Model
fig1 = go.Figure()
# Scatter plot of simulated data points
fig1.add_trace(go.Scatter(x=df_additive[df_additive['Binary Predictor'] == 0]['Continuous Predictor'],
                          y=df_additive[df_additive['Binary Predictor'] == 0]['Response'],
                          mode='markers', name='Binary Predictor = 0', marker=dict(color='blue')))
fig1.add_trace(go.Scatter(x=df_additive[df_additive['Binary Predictor'] == 1]['Continuous Predictor'],
                          y=df_additive[df_additive['Binary Predictor'] == 1]['Response'],
                          mode='markers', name='Binary Predictor = 1', marker=dict(color='red')))
# Add best fit lines
fig1.add_trace(go.Scatter(x=continuous_variable, y=fitted_line_additive_0, mode='lines',
                          name='Additive Model Fit (Binary = 0)', line=dict(color='blue', dash='dash')))
fig1.add_trace(go.Scatter(x=continuous_variable, y=fitted_line_additive_1, mode='lines',
                          name='Additive Model Fit (Binary = 1)', line=dict(color='red', dash='dash')))
fig1.update_layout(title="Additive Model: Continuous + Binary Predictor",
                   xaxis_title="Continuous Predictor", yaxis_title="Response")

# Plot Synergistic Model
fig2 = go.Figure()
# Scatter plot of simulated data points
fig2.add_trace(go.Scatter(x=df_synergistic[df_synergistic['Binary Predictor'] == 0]['Continuous Predictor'],
                          y=df_synergistic[df_synergistic['Binary Predictor'] == 0]['Response'],
                          mode='markers', name='Binary Predictor = 0', marker=dict(color='blue')))
fig2.add_trace(go.Scatter(x=df_synergistic[df_synergistic['Binary Predictor'] == 1]['Continuous Predictor'],
                          y=df_synergistic[df_synergistic['Binary Predictor'] == 1]['Response'],
                          mode='markers', name='Binary Predictor = 1', marker=dict(color='red')))
# Add best fit lines
fig2.add_trace(go.Scatter(x=continuous_variable, y=fitted_line_synergistic_0, mode='lines',
                          name='Synergistic Model Fit (Binary = 0)', line=dict(color='blue', dash='dash')))
fig2.add_trace(go.Scatter(x=continuous_variable, y=fitted_line_synergistic_1, mode='lines',
                          name='Synergistic Model Fit (Binary = 1)', line=dict(color='red', dash='dash')))
fig2.update_layout(title="Synergistic Model: Continuous * Binary Predictor",
                   xaxis_title="Continuous Predictor", yaxis_title="Response")

# Show plots
fig1.show(renderer = 'png')
fig2.show(renderer = 'png')
```

The principles behind the progression of the models can be summarized as follows:

1. Start Simple and Build Up Complexity
   - Rationale: Begin with a basic model to establish a baseline and understand the core relationships between the dependent variable (HP) and a few key predictors (e.g., `Attack` and `Defense` in `model3`).
   - Principle: Simplicity first ‚Äì Simple models are easier to interpret and help identify the most influential predictors before adding more complexity.

2. Incorporate Interactions for Better Accuracy
   - Rationale: As the initial model reveals the basic relationships, the next step is to explore interaction effects (e.g., `Attack * Defense`) to capture how the effect of one predictor changes depending on the values of another. This helps the model account for more complex, real-world relationships between predictors.
   - Principle: Model complexity should increase only when necessary ‚Äì Add complexity (like interactions) when the simpler model fails to capture enough of the data's variability.

3. Simplify and Focus on Key Predictors
   - Rationale: After introducing interactions, the next step is to simplify the model by removing unnecessary terms that might lead to overfitting (e.g., higher-order interactions). Instead, focus on the most relevant variables and predictors that have the strongest relationship with the outcome (e.g., `Type 1`, `Generation` indicators).
   - Principle: Occam's Razor ‚Äì A simpler model with fewer predictors is often more interpretable and generalizes better, reducing the risk of overfitting.

4. Reintroduce Complexities Only If Needed
   - Rationale: After simplifying, if the predictive accuracy does not improve or starts to degrade, reintroduce interactions between predictors to capture more intricate relationships. This can help improve model performance, especially if the simpler models missed important patterns.
   - Principle: Balance between complexity and generalization ‚Äì Adding complexity should be a measured decision aimed at improving prediction, not just fitting the training data more closely.

5. Handle Multicollinearity and Improve Stability

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split

# Define the model specification (adjust based on your previous models)
model_linear_form = 'HP ~ Attack + Defense + Speed + Legendary + Q("Sp. Def") + Q("Sp. Atk")'

# Initialize lists to store the performance metrics
in_sample_r_squared = []
out_sample_r_squared = []

# Set the number of iterations (e.g., 100 for robust testing)
num_iterations = 100

for i in range(num_iterations):
    # Split the data (train-test) randomly without fixing the seed each time
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)
    
    # Define the model
    model_spec = smf.ols(formula=model_linear_form, data=pokeaman_train)
    
    # Fit the model
    model_fit = model_spec.fit()
    
    # In-sample R-squared (on training data)
    in_sample_r_squared.append(model_fit.rsquared)
    
    # Out-of-sample R-squared (on test data)
    yhat = model_fit.predict(pokeaman_test)
    y = pokeaman_test.HP
    out_sample_r_squared.append(np.corrcoef(y, yhat)[0, 1] ** 2)

# Convert lists to arrays for easier analysis
in_sample_r_squared = np.array(in_sample_r_squared)
out_sample_r_squared = np.array(out_sample_r_squared)

# Visualization
plt.figure(figsize=(12, 6))
plt.plot(in_sample_r_squared, label="In-sample R-squared", color='blue', alpha=0.7)
plt.plot(out_sample_r_squared, label="Out-of-sample R-squared", color='red', alpha=0.7)
plt.title("Model Performance: In-sample vs. Out-of-sample R-squared over 100 Iterations")
plt.xlabel("Iteration")
plt.ylabel("R-squared")
plt.legend()
plt.show()
```

Explanation of the Code:
1. **Model Definition**: The model formula (`model_linear_form`) is specified with relevant features such as `Attack`, `Defense`, `Speed`, and other categorical features like `Legendary` and `Type`.
2. **Random Train-Test Split**: In each iteration, the dataset is split randomly into training and test sets. Since `np.random.seed(130)` is **not** fixed within the loop, the splits will differ in each iteration, reflecting real-world randomness in data partitioning.
3. **Model Fitting**: For each random split, the model is fitted to the training data (`pokeaman_train`).
4. **Performance Metrics**:
   - **In-sample R-squared**: This is the R-squared value on the training data, which indicates how well the model explains the variance in the training set.
   - **Out-of-sample R-squared**: This is the R-squared value on the test data (`pokeaman_test`), which gives an indication of how well the model generalizes to unseen data.
5. **Collecting Results**: Both metrics are collected in lists during each iteration.
6. **Visualization**: The results are visualized with two lines, one representing in-sample R-squared and the other representing out-of-sample R-squared, across 100 iterations.

Meaning of Results:
- **In-sample R-squared**: This value reflects how well the model fits the training data. It typically shows a higher value, as the model is optimized for the training data.
- **Out-of-sample R-squared**: This value is more important as it reflects how well the model generalizes to new, unseen data. A significant difference between in-sample and out-of-sample R-squared suggests **overfitting**, where the model performs well on the training data but fails to generalize to the test data.
  
The **purpose of this demonstration** is to illustrate how model performance can vary across multiple train-test splits and how the generalizability of a model can be assessed by comparing in-sample and out-of-sample performance. By visualizing this across many iterations, you get a sense of how stable the model's performance is and whether it is overly tuned to the training data or if it can generalize well to unseen data.
