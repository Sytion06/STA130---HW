The symbol 𝑥ᵢ represents the individual data points or values within the sample that are used to compute summary statistics for that sample.

The symbol 𝑥̄ (x-bar) represents the sample mean, which is calculated from the individual sample data points (𝑥ᵢ). It summarizes the sample but does not directly reflect the entire population.

The symbol 𝜇 represents the population mean, which is the actual average value of the entire population. In hypothesis testing, this is the value we are trying to make inferences about based on the sample data.

The symbol 𝜇₀ represents the hypothesized population mean under the null hypothesis (H₀). It is the assumed population mean if the null hypothesis is true.

The statement from the pre-lecture video implies that when we calculate statistics like 𝑥̄ from our sample, the results also provide information about the population mean 𝜇. The goal of hypothesis testing is to use the sample statistic (𝑥̄) to make inferences about the population parameter (𝜇), testing whether 𝜇 is likely to equal the hypothesized value 𝜇₀.

The p-value is the probability of obtaining results as or more extreme than the observed data, assuming that the null hypothesis is true. Therefore, by imagining a world where the null hypothesis is true, we can simulate the conditions where there is no effect. 

Under this simulated world where the null hypothesis is true, we can create a distrbution of results that we'd expect to see from this simulated world. Then, we take the actual data we collected and compare it to the simulated "null world" data. The p-value therefore tells us the probability of observing data as or more extreme as ours in this hypothetical world where the null hypothesis is true.

The p-value represents the probability of obtaining a test statistic as extreme or more extreme than the one observed, assuming the null hypothesis is true. A high p-value means the observed data is consistent with what we would expect under the null hypothesis. A low p-value suggests that the observed data is very unusual or unlikely if the null hypothesis were true.

Therefore, the smaller the p-value, the less likely it is that the observed data would occur if the null hypothesis were correct. This makes the null hypothesis seem increasingly unreasonable or "ridiculous."

For example, imagine flipping a coin 100 times. If the coin is fair (the null hypothesis), we'd expect to get roughly 50 heads, assuming the null hypothesis is true. However, if we observe 99 heads and 1 tail, this would be highly unexpected under the null hypothesis. Such a scenario would yield a very small p-value, suggesting that it is extremely unlikely that the null hypothesis holds for this data. Therefore, we would conclude that the null hypothesis appears "ridiculous."

### 5. Güntürkün (2003) recorded how kissing couples tilt their heads. 80 out of 124 couples, or 64.5% tilted their heads to the right. Simulate a **p-value** using a "50/50 coin-flipping" model for the assumption of the **null hypothesis** $H_0$ that the population of humans don't have left or right head tilt tendencies when kissing, and use the table below to determine the level of evidence we have against $H_0$. <br>

|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|

![Rodin's sculpture, "The Kiss"
](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d1/Rodin_-_Le_Baiser_06.jpg/409px-Rodin_-_Le_Baiser_06.jpg)


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(130)

# Parameters
n = 124  # Total couples
k_observed = 80  # Observed right tilts
num_simulations = 10000  # Number of simulations

# Simulate coin flips (0 for left tilt, 1 for right tilt)
simulated_tilts = np.random.binomial(n, 0.5, num_simulations)

# Calculate the p-value
p_value = np.mean(simulated_tilts >= k_observed)

# Output the p-value
print(f"Simulated p-value: {p_value:.4f}")

# Optional: Visualize the distribution of simulated right tilts
plt.hist(simulated_tilts, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(k_observed, color='red', linestyle='dashed', linewidth=2, label='Observed Right Tilts')
plt.title('Distribution of Simulated Right Tilts')
plt.xlabel('Number of Right Tilts')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Interpret the p-value
if p_value < 0.001:
    evidence = "very strong evidence against the null hypothesis"
elif 0.001 <= p_value < 0.01:
    evidence = "strong evidence against the null hypothesis"
elif 0.01 <= p_value < 0.05:
    evidence = "moderate evidence against the null hypothesis"
elif 0.05 <= p_value < 0.1:
    evidence = "weak evidence against the null hypothesis"
else:
    evidence = "no evidence against the null hypothesis"

print(f"The p-value indicates: {evidence}.")
```

A smaller p-value cannot definitively prove that the null hypothesis is false, nor can a p-value prove anything with absolute certainty. A p-value simply quantifies how likely it is to observe the data assuming the null hypothesis is true. A high or low p-value only suggests whether or not there is enough evidence to reject the null hypothesis. Both my experiment and the original experiment with Fisher and Dr. Muriel Bristol are designed to test whether someone can distinguish between two conditions (tea first vs. milk first) better than random guessing. The key difference is that Fisher’s experiment involves a small, exact test scenario, while my experiment uses a larger sample and a simulation-based approach for estimating the p-value. Both approaches aim to reject the null hypothesis of random guessing based on the observed outcomes.

Null hypothesis: students are randomly guessing with a 50/50 percent chance of getting the coorect answer. 

Alternative hypothesis: students have some ability to distinguish the order of tea and milk. 

```python
np.random.seed(1)  # make simulation reproducible
number_of_simulations = 10000  # experiment with 10000 simulations
n = 80  # sample size
IncreaseProportionSimulations_underH0random = np.zeros(number_of_simulations)

#assuming H0 (Students are randomly guessing) is true 
for i in range(number_of_simulations):
    random = np.random.choice([0,1], size=n, replace=True)
    IncreaseProportionSimulations_underH0random[i] = random.mean()
    
population_parameter_value_under_H0 = 0.5

observed_test_statistic = 49/80
simulated_test_statistics = IncreaseProportionSimulations_underH0random

# two-tailed test
SimTestStats_as_or_more_extreme_than_ObsTestStat = \
    abs(simulated_test_statistics - population_parameter_value_under_H0) >= \
    abs(observed_test_statistic - population_parameter_value_under_H0) 
    

pd.DataFrame({'(Simulated) Statistic': simulated_test_statistics,
              '>= '+str(observed_test_statistic)+" ?": ['>= '+str(observed_test_statistic)+" ?"]*number_of_simulations, 
              '"as or more extreme"?': SimTestStats_as_or_more_extreme_than_ObsTestStat})
```

```python
p_value = (SimTestStats_as_or_more_extreme_than_ObsTestStat).sum() / number_of_simulations
print("Number of Simulations: ", number_of_simulations, "\n\n",
      "Number of simulated statistics (under HO)\n",
      'that are "as or more extreme" than the observed statistic: ',
      SimTestStats_as_or_more_extreme_than_ObsTestStat.sum(), "\n\n",
      'p-value\n(= simulations "as or more extreme" / total simulations): ', p_value, sep="")
```

In the above code, I assumed that the null hypothesis is true and ran the simulation 10,000 times to generate 10,000 sets of 80 random guesses, each with a 50/50 probability of guessing correctly. With each simulation, I compared the observed statistic (49 correct guesses out of 80) to the simulated statistic under random guessing, using a two-tailed analysis to determine if it is extreme or not.

Finally, I calculated the p-value by dividing the number of times the simulation test statistic is more extreme than the observed test statistic by the total number of simulations.

The p-value I obtained is 0.0439, which means that, assuming the null hypothesis is true (i.e., the students are just randomly guessing), the probability of observing a result as extreme as 49 correct guesses out of 80 (or more extreme) is about 4.39%. This suggests that the observed result (49 correct out of 80) is very unlikely to have occurred by random guessing alone. Therefore, we can reject the null hypothesis since the p-value is less than the significance level of 0.05.

### Ethical Professionalism Considerations
    
Using p-values and hypothesis testing appropriately is an important ethical and professional responsibility of anyone doing data analysis. Actually, there is quite the quiet Contra-Versy (or is it Con-TROV-ersy?) around p-values. First, on a general level, it seems quite clear that p-values and hypothesis testing methodologies MUST play some ongoing contributing role in the so-called "replication crisis" rampantly afflicting mordern science; namely, "significant findings" made in scientific studies are not able to be reproduced by future studies at an alarming rate; and, this whole paradigm of "significant findings" is based on p-values and hypothesis testing... so, something's going on with this methodology in some way...
    
More specifically however, p-values are themselves quite problematic. To see this, just briefly consider the following article titles...

- [Why are p-values controversial?](https://www.tandfonline.com/doi/full/10.1080/00031305.2016.1277161) 
- [What a nerdy debate about p-values shows about science and how to fix it](https://www.vox.com/science-and-health/2017/7/31/16021654/p-values-statistical-significance-redefine-0005)
- [The reign of the p-value is over: what alternative analyses could we employ to fill the power vacuum?](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0174)
- [Scientists rise up against statistical significance](https://www.nature.com/articles/d41586-019-00857-9)
- [Statistics experts urge scientists to rethink the p-value](https://www.spectrumnews.org/news/statistics-experts-urge-scientists-rethink-p-value)

While the issues here are relatively advanced and subtle (as introduced [here](https://www2.stat.duke.edu/~berger/p-values.html), presented [here](https://www.jarad.me/courses/stat587Eng/slides/Inference/I06-Pvalues/why_pvalues_dont_mean_what_you_think_they_mean.pdf), and demonstrated using simulation [here](https://jaradniemi.shinyapps.io/pvalue/)), the problem essentially comes down to the fact that most scientists (or just people) don't know how to really interpret the numeric value of a p-value. There are therefore two current proposed solutions to address this challenge.
    
1. Just interpreting p-values using the follwing table (which really isn't that hard, so it's surprising that this solution isn't more broadly adopted...)
    
|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    

2. Only do **hypothesis testing** on the basis of confidence intervals, not **p-values** (which might be the best solution wherever doing so is a realistic, convenient  possibility...)

With this quite broad introductory context in mind, what does your favorite ChatBot thinks about the following statements? 
    
1. Hypothesis testing is not a "mathematical proof"<br><br>

    1. We do not prove $H_0$ false, we instead give evidence against the $H_0$: "We reject the null hypothesis with a p-value of XYZ, meaning we have ABC evidence against the null hypothesis"
    2. We do not prove $H_0$ is true, we instead do not have evidence to reject $H_0$: "We fail to reject the null hypothesis with a p-value of XYZ"<br><br>

2. Implying that a "non-significant result" means there is "no effect" misleads an audience because this may in actual fact simply indicate that there was insufficient evidence to reject the null hypothesis. So this therefore overlooks the possibility of sample size limitations, or Type II errors (which means a test incorrectly concludes that there is no effect or difference when, in fact, there is one).
