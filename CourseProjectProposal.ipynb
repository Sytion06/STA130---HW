      - Social Interaction Frequency:
           - CONNECTION_preferred_frequency_one_on_one
           - CONNECTION_preferred_frequency_eating_out
           - CONNECTION_preferred_frequency_hang_out
           - CONNECTION_preferred_frequency_eating_in
           - CONNECTION_preferred_frequency_party
           - CONNECTION_preferred_frequency_event
           - CONNECTION_preferred_frequency_sports
           - CONNECTION_preferred_frequency_meet_new
           
      - Mental Health:
           - WORK_feel_stress
           - WELLNESS_self_rated_mental_health
           
      - Life Satisfaction:
           - WELLNESS_life_satisfaction
           - WELLNESS_subjective_happiness_scale_happy
   
   - Analysis:
      - Visualization: Scatterplots and boxplots to illustrate the distribution of mental health indicators and life satisfaction ratings across social interaction frequencies.
      - Correlation Analysis: Calculate the correlation between interaction frequency and life satisfaction/mental wellness.
      - Regression Analysis: Simple linear regression to examine if higher interaction frequency can predict better mental health or life satisfaction, controlling for other factors if available.
      - In the context of simple linear regression, we can test the null hypothesis of "no linear association" between variables. Which involves simulating a sampling distribution and calculating a p-value to assess the evidence against the null hypothesis.
   
   - Hypothesis and Expected Results:
      - Hypothesis: Increased social interaction frequency is positively correlated with mental health and life satisfaction.
      - Relevance
          - Correlation analysis can reveal the strength and direction of the linear association between social interaction frequency and well-being variables. 
          - Simple linear regression can take this exploration a step further. If a statistically significant positive slope coefficient is observed for social interaction frequency when predicting mental health and life satisfaction, it suggests that, "on average," increased frequency of interaction with others is associated with improved self-rated mental health.
          - Hypothesis testing can provide evidence supporting a relationship between the frequency of social interactions and mental health as well as life satisfaction
          
### 2. **Research Question**: Is there an association between community engagement and levels of stress and anxiety?

   - Potential Variables to Explore:
      - Community Engagement:
          - CONNECTION_activities_community_p3m
          - CONNECTION_activities_meeting_organization_p3m
          - CONNECTION_activities_helped_p3m

      - Stress and Anxiety Levels:
          - WORK_feel_stress
          - WELLNESS_maslach_burnout_working_with_people
   
   - Analysis:
      - Visualization: Histograms to show the distribution of stress for participants with varying levels of community engagement. Boxplots comparing stress levels between those who engage frequently versus rarely or never.
      -  Calculate the correlation between each community engagement variable (e.g., CONNECTION_activities_community_p3m) and each stress/anxiety variable (e.g., WORK_feel_stress). This will reveal the strength and direction of the linear relationships between them.
      - Construct simple linear regression models with a stress/anxiety variable as the outcome and a community engagement variable as the predictor. This allows for a more in-depth analysis.
   
   - Hypothesis and Expected Results:
      - Hypothesis: Higher community engagement is associated with lower reported stress and anxiety.
      - Relevance of Analysis: 
          - Correlation coefficients will provide a preliminary understanding of the strength and direction of associations. Negative correlations would suggest that higher community engagement is associated with lower stress/anxiety.
          - Simple linear regression will quantify the "on average" change in stress/anxiety levels for each unit increase in community engagement. A negative slope coefficient would support the hypothesis that greater community engagement is associated with reduced stress/anxiety.
      
### 3. **Research Question**: How does the level of online versus in-person social interaction impact loneliness among individuals?

   - Potential Variables to Explore:
      - Online Interaction Frequency: 
          - CONNECTION_activities_onlinegames_p3m
          - CONNECTION_activities_discussion_group_p3m
          - CONNECTION_close_friends_num_meet_online
      - In-Person Interaction Frequency:
           - CONNECTION_preferred_frequency_eating_out
           - CONNECTION_preferred_frequency_hang_out
           - CONNECTION_preferred_frequency_eating_in
           - CONNECTION_preferred_frequency_party
           - CONNECTION_preferred_frequency_event
           - CONNECTION_preferred_frequency_sports
           - CONNECTION_preferred_frequency_meet_new
      - Loneliness Indicator (e.g., self-reported loneliness scale)
           - LONELY_ucla_loneliness_scale_companionship
           - LONELY_ucla_loneliness_scale_left_out
           - LONELY_ucla_loneliness_scale_isolated
           - LONELY_dejong_emotional_social_loneliness_scale_emptiness
   
   - Analysis:
      - Visualization: Dual bar charts showing levels of loneliness for varying frequencies of online versus in-person interactions.
      - Correlation Analysis: Begin by calculating the correlation between each online interaction variable and each loneliness indicator. Repeat this process for each in-person interaction variable. This will provide an initial understanding of the strength and direction of the linear relationships between these variables.
      - Linear Regression: Construct a multiple linear regression model with a loneliness indicator as the outcome variable and both online and in-person interaction variables as predictors. 
   
   - Hypothesis and Expected Results:
      - Hypothesis: Individuals with higher in-person interaction frequencies report lower loneliness scores than those relying more on online interactions.
      - Relevance of Analysis: 
          - Correlation coefficients will show an initial glimpse into the associations between interaction modes and loneliness. Negative correlations with in-person interaction would suggest that individuals engaging more in face-to-face interactions tend to experience less loneliness. The findings for online interaction might be less straightforward, potentially highlighting differences in the impact of various online activities on loneliness. 
          - Linear regression will enable a more nuanced analysis by isolating the effects of online and in-person interactions while controlling for each other. A larger negative slope coefficient for in-person interaction would support the hypothesis that it plays a more significant role in reducing loneliness.
